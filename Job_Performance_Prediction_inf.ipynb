```python
# Import libraries and the model
import pandas as pd
import numpy as np
import pickle
import json
```

```python
# Load kembali preprocessor dan model
with open("preprocessor.pkl", "rb") as file:
    loaded_preprocessor = pickle.load(file)

with open("best_model.pkl", "rb") as file:
    loaded_model = pickle.load(file)

print(" Model dan Preprocessor berhasil dimuat!")
```

```python
# Buat DataFrame untuk data inference
data_inference = pd.DataFrame({
    'Age': [25, 38, 50],
    'Department': ['Sales', 'IT', 'Marketing'],
    'Job_Level': ['Entry', 'Senior', 'Manager'],
    'Years_at_Company': [1, 13, 18],
    'Monthly_Hours_Worked': [160, 160, 160],
    'Remote_Work': ['No', 'Yes', 'Hybrid'],
    'Meetings_per_Week': [1, 3, 6],
    'Tasks_Completed_Per_Day': [6, 2, 5],
    'Overtime_Hours_Per_Week': [10, 4, 3],
    'Job_Satisfaction' : [4, 7, 8],
    'Work_Life_Balance': ['Poor', 'Average', 'Good'],
    'Annual_Salary': [48000, 65000, 90000],
    'Absences_Per_Year': [10, 13, 8]
})
```

```python
# Preprocessing pada Data Inference
data_inference_processed = loaded_preprocessor.transform(data_inference)

# Prediksi menggunakan model yang telah disimpan
predictions = loaded_model.predict(data_inference_processed)

# Buat DataFrame hasil prediksi
df_result = data_inference.copy()
df_result["Predicted_Class"] = predictions

df_result
```

*The first employee* (index 0) is predicted to be in class 0 (low productivity group) while the second and third employees (index 1 and 2) are predicted to be in class 2 (high productivity group).<bf><br>

Insight: <br>
* Employees with poor work-life balance and high workload tend to be in class 0, which may indicate higher job dissatisfaction or risk of turnover.
* Employees with higher salaries, better work-life balance, and more regular workload are more likely to be in class 2, which may indicate higher job satisfaction and better employee retention.
* The remote work factor does not directly determine the class, as there is a difference between the second and third individuals who are both in class 2, but with different working conditions (remote and hybrid).
